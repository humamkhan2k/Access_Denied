```
train_df = train_df.fillna(train_df.mean())
lead_df = lead_df.fillna(lead_df.mean())
eval_df = eval_df.fillna(eval_df.mean())
```

```
data = [train_df, lead_df, eval_df]
for dataset in data:
    dataset['VAR18'] = dataset['VAR18'].astype(float)
    dataset['VAR19'] = dataset['VAR19'].astype(float)
```

```
train_df = train_df.drop(['VAR14'], axis=1)
lead_df = lead_df.drop(['VAR14'], axis = 1)
eval_df = eval_df.drop(['VAR14'], axis = 1)
```

```
value = {"Low": 0.0, "Medium": 1.0, "High": 2.0}
data = [train_df]

for dataset in data:
    dataset['VAR21'] = dataset['VAR21'].map(value)
```

```
# Splting into train and test data

X = train_df.iloc[:, 0:19]
Y = train_df.loc[:, 'VAR21']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```
# Normalizing X_train

names = X_train.columns
scaler = preprocessing.StandardScaler()

scaled_df = scaler.fit_transform(X_train)
scaled_df * 10
scaled_df = pd.DataFrame(scaled_df, columns = names)
X_train = scaled_df.round(4)
```

```
# Normalizing X_test

names = X_test.columns
scaler = preprocessing.StandardScaler()

scaled_testdf = scaler.fit_transform(X_test)
scaled_testdf * 10
scaled_testdf = pd.DataFrame(scaled_testdf, columns = names)
X_test = scaled_testdf.round(4)
```

```
# KNN Model

knn = KNeighborsClassifier(n_neighbors=80)
knn.fit(X_train, Y_train)

print(knn)

Y_pred = knn.predict(X_test)

print("Accuracy: ",  metrics.accuracy_score(Y_test, Y_pred))
```

```
# XGBoost Model

model1 = XGBClassifier()
model1.fit(X_train, Y_train)

print(model1)

Y_pred = model1.predict(X_test)
print(Y_pred)

print("Accuracy: ",  metrics.accuracy_score(Y_test, Y_pred))
```

```
# Naive Bayes Classifier

model2 = GaussianNB()
model2.fit(X_train, Y_train)

print(model2)

Y_pred = model2.predict(X_test)

print("Accuracy: ",  metrics.accuracy_score(Y_test, Y_pred))
```

```
# Random Forest Classifier Model

model3 = RandomForestClassifier(n_estimators = 3)
model3.fit(X_train, Y_train)

print(model3)

Y_pred = model3.predict(X_test)
print(Y_pred)

print("Accuracy: ",  metrics.accuracy_score(Y_test, Y_pred))
```

```
# Decision Tree Model using gini

model4 = DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth=4, min_samples_leaf=5)
model4.fit(X_train, Y_train)

print(model4)

Y_pred = model4.predict(X_test)

print("Accuracy: ",  metrics.accuracy_score(Y_test, Y_pred))
```

```
# Decision Tree Model using entropy

model5 = DecisionTreeClassifier(criterion = "entropy", random_state = 100, max_depth=6, min_samples_leaf=5)
model5.fit(X_train, Y_train)

print(model5)

Y_pred = model5.predict(X_test)

print("Accuracy: ",  metrics.accuracy_score(Y_test, Y_pred))
```

```
# SGD Classifier

model6 = linear_model.SGDClassifier(max_iter=2000, tol=1e-3)
model6.fit(X_train, Y_train)

print(model6)

Y_pred = model6.predict(X_test)

print("Accuracy: ",  metrics.accuracy_score(Y_test, Y_pred))
```

```
# Logistic Regression Classifier solver = 'saga' or 'sag' or 'newton-cg' or 'lbfgs'

model7 = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial')
model7.fit(X_train, Y_train)

print(model7)

Y_pred = model7.predict(X_test)

print("Accuracy: ",  metrics.accuracy_score(Y_test, Y_pred))

Y_pred
```

```
# Neural Network Model

model = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(6, 4, 2), activation='tanh', max_iter=300)
model.fit(X_train, Y_train)

print(model)

Y_pred = model.predict(X_test)

print("Accuracy: ",  metrics.accuracy_score(Y_test, Y_pred))

Y_pred
```
